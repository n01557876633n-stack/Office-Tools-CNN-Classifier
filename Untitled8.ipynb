```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# إعدادات التكليف: Resize لـ 32x32
data_transforms = transforms.Compose([
    transforms.Resize((32, 32)),
    transforms.RandomHorizontalFlip(), # يقلب الصورة يمين وشمال
    transforms.RandomRotation(10),      # يميل الصورة 10 درجات
    transforms.ColorJitter(brightness=0.2), # يغير في الإضاءة شوية
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

train_path = '/content/drive/MyDrive/Office_Tools_Project/train'
test_path = '/content/drive/MyDrive/Office_Tools_Project/test'

train_dataset = datasets.ImageFolder(root=train_path, transform=data_transforms)
test_dataset = datasets.ImageFolder(root=test_path, transform=data_transforms)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

print(f"Loaded {len(train_dataset)} training images.")
print(f"Classes: {train_dataset.classes}")
```

```python
class CNNModel(nn.Module):
    def __init__(self):
        super(CNNModel, self).__init__()
        # Block 1: in=3, out=16
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, padding=1)
        # Block 2: in=16, out=32
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, padding=1)
        # Block 3: in=32, out=64
        self.conv3 = nn.Conv2d(32, 64, kernel_size=3, padding=1)

        self.pool = nn.MaxPool2d(2, 2)

        # FC Layers
        self.fc1 = nn.Linear(64 * 4 * 4, 128)
        self.fc2 = nn.Linear(128, 3) # عندنا 3 أصناف

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x))) # Conv 1 -> ReLU -> Pool
        x = self.pool(F.relu(self.conv2(x))) # Conv 2 -> ReLU -> Pool
        x = self.pool(F.relu(self.conv3(x))) # Conv 3 -> ReLU -> Pool
        x = x.view(-1, 64 * 4 * 4)           # Flatten
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

model = CNNModel()
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

print("Starting Training...")
for epoch in range(50):
    running_loss = 0.0
    for images, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    print(f"Epoch [{epoch+1}/50], Loss: {running_loss/len(train_loader):.4f}")
print("Training Finished!")
```

```python
correct = 0
total = 0
model.eval()
with torch.no_grad():
    for images, labels in test_loader:
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f'Final Accuracy: {100 * correct / total:.2f}%')
```

```python
from PIL import Image
import matplotlib.pyplot as plt

def predict_image(image_path):
    # 1. تحميل الصورة
    img = Image.open(image_path)

    # 2. تحويل الصورة لنفس التنسيق اللي الموديل اتدرب عليه
    transform = transforms.Compose([
        transforms.Resize((32, 32)),
        transforms.ToTensor(),
        transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
    ])

    # تجهيز الصورة (إضافة بُعد الـ Batch)
    img_tensor = transform(img).unsqueeze(0)

    # 3. التوقع (Prediction)
    model.eval()
    with torch.no_grad():
        output = model(img_tensor)
        _, predicted = torch.max(output, 1)

    # 4. عرض النتيجة
    class_names = train_dataset.classes # [Pen, Scissors, Stapler]
    print(f"Prediction: {class_names[predicted.item()]}")
    plt.imshow(img)
    plt.axis('off')
    plt.show()

# جربي الكود هنا! حطي مسار أي صورة من الدرايف أو ارفعي صورة جديدة
# predict_image('/content/any_image.jpg')
```
